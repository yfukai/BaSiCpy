```python
import numpy as np
from matplotlib import pyplot as plt
from os import path
from skimage import io
```

```python
images_distorted_wo_darkfield = np.load("images_distorted_wo_darkfield_with_noise.npy")
flatfield_true = np.load("flatfield_true.npy")
weights = io.imread("weights.tif")
images_distorted_wo_darkfield.shape
```

```python
flatfield_half = (flatfield_true-1)/2+1
```

```python
im = images_distorted_wo_darkfield[1,-1]
w = weights[1,-1]
im_half_corrected = images_distorted_wo_darkfield[1,-1]/flatfield_half
im_corrected = images_distorted_wo_darkfield[1,-1]/flatfield_true
```

```python
plt.subplot(131)
plt.imshow(im, cmap='gray')
plt.subplot(132)
plt.imshow(im_half_corrected, cmap='gray')
plt.subplot(133)
plt.imshow(im_corrected, cmap='gray')
```

```python
plt.hist(im.ravel(),bins=512);
plt.hist(im_half_corrected.ravel(),bins=512);
plt.hist(im_corrected.ravel(),bins=512);
```
